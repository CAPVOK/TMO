# Анализ и прогнозирование временного ряда

```python
# Импорт необходимых библиотек
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.datasets import sunspots
data = sunspots.load_pandas().data['SUNACTIVITY']
```

```python
# Визуализация временного ряда
plt.figure(figsize=(10, 6))
plt.plot(data)
plt.title('Sunspots Time Series')
plt.xlabel('Time')
plt.ylabel('Sunspots')
plt.grid(True)
plt.show()
```

```python
# Разделение временного ряда на обучающую и тестовую выборки
train_size = int(len(data) * 0.8)
train, test = data[0:train_size], data[train_size:len(data)]
```

После подготовки перейдем к прогнозированию временного ряда

```python
# ARIMA модель
model_arima = ARIMA(train, order=(5,1,0))
model_arima_fit = model_arima.fit()
forecast_arima = model_arima_fit.forecast(steps=len(test), index=test.index)[0]

```
